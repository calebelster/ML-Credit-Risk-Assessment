```python
import numpy as np
import pandas as pd

np.random.seed(42)  # for reproducibility

n = 100

home_ownership_options = ["RENT", "OWN", "MORTGAGE", "OTHER"]
loan_intent_options = [
    "PERSONAL", "EDUCATION", "MEDICAL",
    "VENTURE", "HOMEIMPROVEMENT", "DEBTCONSOLIDATION"
]
default_options = ["Y", "N"]

# Age first
person_age = np.random.randint(18, 76, size=n)

# Emp length: 0 up to (age - 1), but also cap at 40 if you like
person_emp_length = np.array([
    np.random.randint(0, min(age, 41)) for age in person_age
])

# Credit history length: 0 up to (age - 1)
cb_person_cred_hist_length = np.array([
    np.random.randint(0, age) for age in person_age
])

data = {
    "person_age": person_age,
    "person_income": np.random.uniform(20_000, 200_000, size=n).round(2),
    "person_emp_length": person_emp_length,
    "loan_amnt": np.random.uniform(1_000, 50_000, size=n).round(2),
    "loan_int_rate": np.random.uniform(3.0, 36.0, size=n).round(2),
    "cb_person_cred_hist_length": cb_person_cred_hist_length,
    "person_home_ownership": np.random.choice(home_ownership_options, size=n),
    "loan_intent": np.random.choice(loan_intent_options, size=n),
    "cb_person_default_on_file": np.random.choice(default_options, size=n, p=[0.2, 0.8]),
}

df = pd.DataFrame(data)
df.to_csv("mock_batch.csv", index=False)
print("Saved mock_batch_applications_age_consistent.csv")
```
